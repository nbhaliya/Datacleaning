```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("loan analysis.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.displot(df["ApplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(df[["ApplicantIncome"]])
```

```python
df["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(df[["ApplicantIncome"]]),columns=["x"])
```

```python
df.head(3)
```

```python
plt.subplot(1,2,1)
plt.title("before")
sns.displot(df["ApplicantIncome_ss"])
plt.subplot(1,2,1)
plt.title("after")

sns.displot(df["ApplicantIncome"])

plt.show()
```

```python
plt.figure(figsize=(12, 5))  # Set figure size

# First subplot (Before)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.title("Before")
sns.histplot(df["ApplicantIncome_ss"], kde=True, bins=30)  # Use histplot instead of displot

# Second subplot (After)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.title("After")
sns.histplot(df["ApplicantIncome"], kde=True, bins=30)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
